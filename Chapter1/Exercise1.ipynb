```python
import numpy as np

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Sigmoid neuron
def sigmoid_neuron(x, w, b):
    z = np.dot(w, x) + b
    return sigmoid(z)

# Input
x = np.array([0.6, 0.9])

# Original weights and bias
w = np.array([1.0, -1.0])
b = -0.2

# Scaling factor (positive constant)
c = 50   # try increasing this

# Output before scaling
output_original = sigmoid_neuron(x, w, b)

# Output after scaling weights and bias
output_scaled = sigmoid_neuron(x, c * w, c * b)

print("Original sigmoid output:", output_original)
print("Scaled sigmoid output:  ", output_scaled)

# Perceptron-like decision
print("Original decision:", int(output_original > 0.5))
print("Scaled decision:  ", int(output_scaled > 0.5))
```
