```python
#updation of wtgs and biases for each mini batches
def update_mini_batch(self,mini_batch,eta):
    # Each min batch will be initialized with 0 for their gradient accumulaters (gradient wtgs and gradient bias ) to note the changes for each one sample in the min batch and update the original weights and bias at last 
    m=len(mini_batch)
    gradient_weight=[ np.zeros(w.shape) for w in self.weights]
    gradient_bias=[ np.zeros(b.shape) for b in self.biases]
    for X,y in mini_batch:
        change_gradient_w,change_gradient_b=self.backpropagation(X,y)
        # summing up the changed and gradient acummulaters
        gradient_weight=[ gw+cgw for gw,cgw in zip(gradient_weight,change_gradient_w)]
        gradient_bias=[ gb+cgb for gb,cgb in zip(gradient_bias,change_gradient_b)]
    # updation of original weight and bias at the end of each mini batch 
    self.weights=[ (w-(eta*gw)/m) for w,gw in zip(self.weights,gradient_weight)]
    self.biases=[ (b-(eta*gb)/m) for b,gb in zip(self.biases,gradient_bias)]
```

Implementation of MNIST using SVM classifier

```python
import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
X,y=fetch_openml('mnist_784',return_X_y=True)
```

```python
X=X/255.0
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.15,random_state=42)
```

```python
model=svm.LinearSVC()
model.fit(X_train,y_train)
```

```python
predict=model.predict(X_test)
crct=np.sum(predict==y_test)
```

```python
print(crct)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predict))
```
